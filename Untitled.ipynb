```python
genre_counts = manga_df['genres'].value_counts()

# Set up the figure and axis
plt.figure(figsize=(10, 6))
ax = sns.barplot(x=genre_counts.index, y=genre_counts.values, palette='viridis')
plt.xticks(rotation=90)  # Rotate x-axis labels for better readability

# Add labels and title
plt.xlabel('Genre')
plt.ylabel('Count')
plt.title('Frequency of Genres in Manga')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
import re

# Extract all unique words within single quotes from the 'genre' column
unique_words = manga_df['genres'].str.extractall(r"'(.*?)'")[0].unique()

# Print the unique words
print(unique_words)
```

```python
word_counts = pd.DataFrame({'Word': unique_words, 'Count': [manga_df['genres'].str.count(fr"'{word}'").sum() for word in unique_words]})

# Sort the DataFrame by count in descending order
word_counts = word_counts.sort_values(by='Count', ascending=False)

# Set up the figure and axis
plt.figure(figsize=(10, 6))
plt.bar(word_counts['Word'], word_counts['Count'], color='skyblue')
plt.xticks(rotation=90)

# Add labels and title
plt.xlabel('Genres')
plt.ylabel('Count')
plt.title('Frequency of genre tags in Manga')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
for word in unique_words:
    manga_df[word] = manga_df['genres'].str.contains(fr"'{word}'", case=False)

# Fill the new Boolean columns with True or False
manga_df = manga_df.astype({word: bool for word in unique_words})
```

```python
manga_df
```

```python
word_counts = {}

# Iterate through the unique words and count the True values in each column
for word in unique_words:
    word_counts[word] = manga_df[word].sum()

# Convert the dictionary to a DataFrame for better visualization
word_counts_df = pd.DataFrame.from_dict(word_counts, orient='index', columns=['Count'])

# Sort the DataFrame by count in descending order
word_counts_df = word_counts_df.sort_values(by='Count', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(word_counts_df.index, word_counts_df['Count'])
plt.xticks(rotation=90)

# Add labels and title
plt.xlabel('Word')
plt.ylabel('Count')
plt.title('Frequency of genre tags in Manga')

# Show the plot
plt.tight_layout()
plt.show()
```

```python
from matplotlib.ticker import ScalarFormatter
import matplotlib.pyplot as plt

word_favorites_sum = {}

# Iterate through the unique words and calculate the sum of 'favorites' for each word
for word in unique_words:
    word_favorites_sum[word] = manga_df[manga_df[word]]['favorites'].sum()

# Convert the dictionary to a DataFrame for better visualization
favorites_sum_df = pd.DataFrame.from_dict(word_favorites_sum, orient='index', columns=['Favorites_Sum'])

# Sort the DataFrame by favorites sum in descending order
favorites_sum_df = favorites_sum_df.sort_values(by='Favorites_Sum', ascending=False)

plt.figure(figsize=(10, 6))
plt.bar(favorites_sum_df.index, favorites_sum_df['Favorites_Sum'])
plt.xticks(rotation=90)

# Add labels and title
plt.xlabel('Genre')
plt.ylabel('Sum of Favorites')
plt.title('Most of Favorited Genres in Manga')

# Use ScalarFormatter to format the y-axis labels
ax = plt.gca()
ax.yaxis.set_major_formatter(ScalarFormatter(useMathText=True))

# Show the plot
plt.tight_layout()
plt.show()
```

I realized the data also had doujinshi and one_shot in the data which drastically effected the accuracy of the information I needed.

doujinshi are basically 'indie' mangas which are usually based on other serializations. the popularity of which greatly varies based on source material and can effectively create duplicate of the source genre or end up in the 'Hentai' genre

one_shot are chapter demos usually submited to publications and then tested to see if they can get fully serialized 

both of these are usually just one chapters that aren't represenative of a full series thus need to be excluded from the data.
